## ***Create the Decision Tree classifier and visualize it graphically.***

### ***Import the required liabrires***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import tree
import sklearn.datasets as datasets
```

### ***Load the iris dataset***

```python
iris=datasets.load_iris()
```

### ***Put all the features into X variable***

```python
X=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
X.head()
```

### ***Taeget values into y variable***

```python
y=iris.target
y
```

### ***Checking if there is any null value***

```python
X.isnull().sum()
```

### ***Describing the features of iris dataset***

```python
X.describe()
```

### ***Infromation about the features of iris dataset***

```python
X.info()
```

### ***Splitting the data into train and test data***

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

### ***Creating an obeject for decision tree***

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

### ***Fitting the train dataset into the model***

```python
model.fit(X_train,y_train)
```

### ***Prediciting the test data***

```python
y_predict=model.predict(X_test)
y_predict
```

### ***Visualizing the Decision Tree***

```python
feature_n=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
class_n=['setosa','versicolor','virginica']
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)
tree.plot_tree(model,feature_names=feature_n,class_names=class_n,filled='True',rounded='True')
```

### ***Accuracy of the model***

```python
model.score(X_test,y_test)
```

## ***The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly. and  we got accuracy of this model to our test data as 0.96 which is 96% accurate enough to use this model***
